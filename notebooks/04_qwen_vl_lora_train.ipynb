# Qwen2.5-VL LoRA Fine-Tuning

This notebook records the Qwen2.5-VL LoRA fine-tuning setup used for fine-grained English label generation.

```python
training_command = """
python scripts/train_qwen_lora.py \
  --model_name_or_path $QWEN_VL_MODEL_PATH \
  --dataset_use trainset \
  --output_dir outputs/qwen_lora/vision_mlp_r64 \
  --bf16 \
  --eval_strategy no \
  --save_strategy steps \
  --save_steps 1000 \
  --logging_steps 1 \
  --per_device_train_batch_size 1 \
  --gradient_accumulation_steps 4 \
  --learning_rate 1e-5 \
  --warmup_ratio 0.03 \
  --lr_scheduler_type cosine \
  --weight_decay 0.01 \
  --lora_enable True \
  --tune_mm_mlp True \
  --tune_mm_llm False \
  --tune_mm_vision True \
  --lora_r 64 \
  --num_train_epochs 0.5 \
  --lora_alpha 16 \
  --lora_dropout 0.05 \
  --max_pixels 313600 \
  --min_pixels 3136 \
  --report_to tensorboard
"""
print(training_command)
```
