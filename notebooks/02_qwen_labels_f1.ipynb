# Qwen2.5-VL Fine-Grained Label Retrieval

This notebook documents the fine-grained retrieval pipeline. Qwen2.5-VL generates image labels, the labels are encoded with Taiyi-CLIP, and retrieval is evaluated in the label-feature space.

```python
from pathlib import Path
import json

metadata_path = Path("../data/metadata/qwen_lora_labels.json")
with metadata_path.open("r", encoding="utf-8") as f:
    data = json.load(f)

print(data["images"][0])
```

## English Label-Generation Prompts

```python
open_label_prompt = """
<image>
Describe the main objects visible in the image.
Return only a JSON-style list of short English object labels, for example:
["chair", "table", "cup"]
"""

target_label_prompt = """
<image>
Identify whether the image contains any of the following target categories:
["chair", "cup", "bicycle", "skateboard", "oven"]
Return only a JSON-style list of matching target categories.
If none are present, return ["none_of_target_categories"].
"""

print(open_label_prompt)
print(target_label_prompt)
```

## Retrieval Evaluation

The generated labels are encoded as text features. The online query is encoded with the same text encoder, then cosine similarity is used for ranking and threshold evaluation.
