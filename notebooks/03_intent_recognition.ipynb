# Intent Recognition Dataset And Qwen Evaluation

This notebook documents the English intent-recognition task used by the multimodal retrieval agent.

```python
intent_generation_prompt = """
Generate user queries for an image retrieval assistant.
Each example must be JSON with the following fields:
{
  "question": "",
  "needs_retrieval": 0 or 1,
  "query_type": "text_to_image" | "image_to_image" | "no_retrieval",
  "category": "chair" | "cup" | "bicycle" | "skateboard" | "oven" | "",
  "search_mode": ["top_k", K] | ["threshold"] | [""]
}

Use diverse natural English phrasing.
If a specific number of images is requested, use ["top_k", K].
If the quantity is vague, use ["threshold"].
"""
print(intent_generation_prompt)
```

```python
from pathlib import Path
import json

intent_path = Path("../data/metadata/gpt_intent.json")
with intent_path.open("r", encoding="utf-8") as f:
    intent_data = json.load(f)

for split, records in intent_data.items():
    print(split, len(records), records[0])
```

## English Qwen Evaluation Prompt

```python
qwen_eval_prompt = """
Classify the user request for an image retrieval system.
Return only valid JSON with these fields:
{
  "question": "<the original user request>",
  "needs_retrieval": 0 or 1,
  "query_type": "text_to_image" | "image_to_image" | "no_retrieval",
  "category": "chair" | "cup" | "bicycle" | "skateboard" | "oven" | "",
  "search_mode": ["top_k", K] | ["threshold"] | [""]
}
Do not add explanations.
"""
print(qwen_eval_prompt)
```
